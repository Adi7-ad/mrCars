# Loading Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# 

# Loading and Understanding the Data

```python
df = pd.read_csv("mtcars.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# 

# Working on data

```python
# Handling Missing Value

df.isnull().sum()
print("Missing Values:\n", df.isnull().sum())
```

```python
# Handling Categorical Data

df["cyl"]= df["cyl"].astype("category")
```

```python
df.info()
```

# 

#Univariate Analysis

```python
# Distribution of MPG

sns.histplot(df["mpg"], bins=30, kde=True)
plt.title("Mileage")
plt.show()
```

```python
# Histogram of Cylinder

sns.countplot(x="cyl", data=df)
plt.title("Cylinder Size wise distribution")
plt.show()
```

```python
# Histogram of HP

sns.histplot(df["hp"], kde=True, bins=30)
plt.title("Distribution of HorsePower")
plt.show()
```

```python
# Histogram of Weight

sns.histplot(x="wt", data=df)
plt.title("Distribution of Weight")
plt.show()
```

# 

#Model-Wise Mileage Distribution


```python
plt.figure(figsize=(12,10))
sns.histplot(data=df, x="mpg", y='model', hue="model", element='step', palette='tab10', bins=30)
plt.title("Model wise Mileage")
plt.xlabel("Model")
plt.ylabel("Mileage")
plt.show()
```

# 

#Multivariante Analysis

```python
# Mileage Per Gallon Vs Cylender 

sns.boxplot(x="cyl", y="mpg", data=df)
plt.title("Cyliner Vs MPG")
plt.show()
```

```python
# Correlation Analysis

corr_mat = df.select_dtypes(include=["number"]).corr()
plt.figure(figsize=(10,5))
sns.heatmap(corr_mat, annot= True, cmap= "coolwarm", fmt= ".1f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show
```

```python
# Feature Engineering 

df["Power_to_Weight"]= df["hp"]/df["wt"]
df
```

```python
# Detecting Outliers

sns.boxplot(x= df["mpg"])
plt.title("Boxplot in MPG")
plt.show()
```

# 

#Visualization Technique 

```python
#built a pair wise scatter plot

sns.pairplot(df[["mpg","hp","wt","am"]])
plt.show
```
